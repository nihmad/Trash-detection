# Exploration du dataset TrashNet

Ce notebook donne un aperçu rapide du dataset : distribution des classes,
détection d'images corrompues et visualisation d'exemples.

Prérequis : avoir placé les images dans `data/raw/` (un dossier par classe).

```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

from src.dataset import build_index, class_distribution, split_dataset
from src.utils import DATA_RAW_DIR
```

```python
index = build_index(DATA_RAW_DIR)
print(f"Nombre total d'images : {len(index)}")
class_distribution(index)
```

```python
import matplotlib.pyplot as plt

class_distribution(index).plot(kind="bar")
plt.title("Distribution des classes")
plt.ylabel("Nombre d'images")
plt.tight_layout()
plt.show()
```

```python
from src.preprocessing import preprocess_image

sample = index.groupby("label").first().reset_index()
fig, axes = plt.subplots(1, len(sample), figsize=(15, 3))
for ax, row in zip(axes, sample.itertuples()):
    ax.imshow(preprocess_image(Path(row.filepath)))
    ax.set_title(row.label)
    ax.axis("off")
plt.show()
```

```python
splits = split_dataset(index)
print(f"train={len(splits.train)}, val={len(splits.val)}, test={len(splits.test)}")
```
